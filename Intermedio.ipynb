```python
def mifuncion(x,y):
    """
    Esta es la ayuda de mifuncion, mifuncion realiza la suma de 2 numeros. 
    """
    suma=x+y    
    return suma
mifuncion(3,2)
#help(mifuncion)
```

En este caso el nombre de la función es *mifuncion* y tiene como entrada dos parámetros x e y. Se pueden realizar funciones sin parámetros y además poner parámetros como defecto o inicializarlos. Por ejemplo:

Las varibales que se declaran en las funciones son variables locales y se encuentran unicamente en el entorno de la función y desaparecen luego de que esta se halla ejecutado. Ejemplo:

```python
b=0.5
def raizcuadrada(a):
    global b
    b=3
    raiz=a**b
    return raiz
raizcuadrada(4)

#Cual es la variable local y la global
```

```python
#Suma, multiplicacion por escalar, producto punto y cruz de vectores
a=np.array([1,2,3])
b=np.array([2,3,4])
print(a+b) #Suma
print(np.dot(a,b)) #Producto punto
print(4*a) # multiplicacion por escalar
print(np.cross(a,b))
print(len(a)) #Longitud o número de componentes de un vector
```

```python
#Suma, Multiplicación, contracción, inversa, transpuesta de matrices
M=np.array([[1,2,3],[4,5,6],[3,2,3]])
N=np.zeros((3,3))
I=np.eye(3)
print(M)
print(N)
print(I)
mult=M@I
print("Multiplicacion de matrices: \n", mult)
print(M*N) #Multiplicacion de cada elemento de las matrices

#Ejercicio: construir la función contracción


invM=np.linalg.inv(M) # Inversa de una matriz
tranM=np.transpose(M) # Transpuesta de una matriz
print(M[0]) # Accediendo a la primera fila de la matriz
print(M[1,1]) # Accediendo al numero ubicado en la fila 2 y la columna 2
```

## Modulo Matplotlib ##
Sirve para crear gráficos, para conocer toda la documentación: ir a https://matplotlib.org/ Se trabajará en este cuaderno con gráficas 2D. 

```python
#
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
```

```python
x=np.linspace(0,6.28,30) #linspace(inicio,fin,numero de puntos)
y1=np.cos(x)
y2=np.sin(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend("12")
plt.ylabel("Valor de funcion")
plt.xlabel("Ang en radianes")
plt.title("Funciones trigonometricas")
plt.grid() #Activar rejilla
plt.show()
```

```python
"""
El cuaderno de notebook permite agregar funcionalidad interactiva mediante el uso de Interact
"""
from ipywidgets import interact

def funciones(freq):
    t=np.linspace(0,6.28,30) #linspace(inicio,fin,numero de puntos)
    y1=np.cos(freq*t)
    y2=np.sin(freq*t)
    plt.plot(x,y1)
    plt.plot(x,y2)
    plt.legend("12")
    plt.ylabel("Valor de funcion")
    plt.xlabel("Ang en radianes")
    plt.title("Funciones trigonometricas")
    plt.grid() #Activar rejilla
    plt.show()
    

interact(funciones,freq=(0.1,4))
```
